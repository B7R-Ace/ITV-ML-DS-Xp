## Data Strategy Challange: Cleaning and preparing data
Part of the challenge aimed at understanding and familiarizing with the dataset.

```python
#importing modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#importing file and checking it
raw_data = pd.read_excel('Copy of DIG_Step_Up_Viewing_Data_Strategy.xlsx')
# raw_data.head()
```

```python
#Removing uneccesary columns to make data more readable/manageable
cleaned_data = raw_data.drop(['user_id', 'stream_id', 'session_start_datetime', 'session_end_datetime', 'programme_id', 'series_number', 'episode_title', 'episode_number', 'schedule_programme_slot_duration'], axis = 1)
#cleaned_data.head()
```

```python
## Seperating Viewers and Non-viewers of ASAF

viewers = cleaned_data[cleaned_data['watched_flag'] == 1]
non_viewers = cleaned_data[cleaned_data['watched_flag'] == 0]

```

## Actual visualisation and analysis starts here:
For more generalized comparisons pie charts are used whereas for accurate feature dependant properties bar graphs are used.

```python
# Comparing number/percentage of viewers to non-viewers

count_n_v = (cleaned_data['watched_flag'] == 0).sum()
count_v = (cleaned_data['watched_flag'] == 1).sum()
y = [count_n_v, count_v]
labels = ['Non-ASAF viewers', 'ASAF Viewers']

# plt.figure(figsize=(5, 500))
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title('A Spy amoung friends viewers vs non-viewers overall data')
```

```python
##Comparing popular stream types for each type of viewer:

stream_types_v = viewers['stream_type'].value_counts()
stream_types_n_v = non_viewers['stream_type'].value_counts()

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20, 15))
fig.suptitle('Stream type comparison for both viewer classes', fontsize='28', y = 0.8)
ax1.pie(stream_types_n_v, labels= stream_types_n_v.index, autopct = '%1.1f%%')
ax2.pie(stream_types_v, labels= stream_types_v.index, autopct = '%1.1f%%')
ax1.set_title('Non-ASAF watchers', fontweight='bold')
ax2.set_title('ASAF watchers', fontweight='bold')
plt.show()
```

```python
## Comparing weekday watching volume to weekend volume for both viewer types

v_weekdata = viewers['is_weekend'].value_counts()
n_v_weekdata = non_viewers['is_weekend'].value_counts()

fig, (axis1, axis2) = plt.subplots(1, 2, figsize=(20, 15))
fig.suptitle('Viewership numbers compared weekend vs weekday', fontsize='22', y=0.8)
axis1.pie(v_weekdata, labels=['Weekday', 'Weekend'], autopct='%1.0f%%')
axis2.pie(n_v_weekdata, labels=['Weekday', 'Weekend'], autopct='%1.0f%%')
plt.subplots_adjust(wspace=0.4, hspace=0.7)
axis1.set_title('ASAF watchers', fontweight='bold')
axis2.set_title('non-ASAF watchers', fontweight='bold')
```

**Interestingly there isn't much difference observed in viewership pattern based on time of the week for these 2 different categories**

**Investigating subplot level general data overiew again. This time looking at genres**

```python
genre_trend_v = ['Action', 'Comedy', 'Drama', 'Fantasy']
asaf_viewers = [150, 200, 180, 120]
non_asaf_viewers = [80, 120, 100, 70]

# Create a figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 15))

# Plot bar graph on the first subplot (ax1)
ax1.bar(genre_trend_v, asaf_viewers, color='orange')
ax1.set_title('Orange: ASAF viewers')
ax1.set_xlabel('Genres')
ax1.set_ylabel('Viewers')

# Plot bar graph on the second subplot (ax2)
ax2.bar(genre_trend_v, non_asaf_viewers, color='blue')
ax2.set_title('Blue: Non-ASAF Viewers')
ax2.set_xlabel('Genres')
ax2.set_ylabel('Viewers')

# Rotate x-axis labels for better readability
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)

# Add spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
